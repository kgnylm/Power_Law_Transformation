```python
"""
Author:Mustafa Kağan Yalım
270201090
"""
import cv2 as cv
from skimage import io 
import matplotlib.pyplot as plt
import numpy as np
```

```python
def plot_image(image_1, image_2, image_3, title_1="Original", title_2="Gray Scale", title_3="Power Law"):


    plt.figure(figsize=(25,25))
    plt.subplot(1, 3, 1)
    plt.imshow(image_1,cmap="gray")
    plt.title(title_1)
    plt.subplot(1, 3, 2)
    plt.imshow(image_2,cmap="gray")
    plt.title(title_2)
    plt.subplot(1, 3, 3)
    plt.imshow(image_3,cmap="gray")
    plt.title(title_3)
    plt.show()

def p_law(image,gamma_rate):
    gamma = np.array(255*(image/255)**gamma_rate,dtype='uint8')
    return gamma
```

```python
img=io.imread("https://www.timeforkids.com/wp-content/uploads/2019/09/final-cover-forest.jpg")
img_gray=cv.cvtColor(img,cv.COLOR_BGR2GRAY)

gamma_rate=5
final_gamma_picture=p_law(img_gray,gamma_rate)

plot_image(img,img_gray,final_gamma_picture,"Original RGB","Gray Scale",f"Power Law: {gamma_rate}")
```
